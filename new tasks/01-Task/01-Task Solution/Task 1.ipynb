```python
import requests
from bs4 import BeautifulSoup
 

url = "http://www.values.com/inspirational-quotes"
# send request to download the data
response = requests.request("GET", url)

data = BeautifulSoup('html.parser')

# parse the downloaded data
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
quotes=[]  # a list to store quotes
  
table = data.find('div', attrs = {'id':'all_quotes'})
  
for row in table.findAll('div', attrs = {'class':'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote = {}
    quote['theme'] = row.h5.text
    quote['url'] = row.a['href']
    quote['img'] = row.img['src']
    quote['lines'] = row.img['alt'].split(" #")[0]
    quote['author'] = row.img['alt'].split(" #")[1]
    quotes.append(quote)
print(quotes)
```

```python
import pandas as pd

# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(quotes)

dataFrame.head()
# save the scraped data as CSV file
dataFrame.to_csv('web_q1_quotes_data.csv', index=False)
```
